```python
 import pandas as pd
```

```python
Project_Data=pd.read_csv("Downloads/Python_SalesData.xlsx - Raw.csv")
```

```python
Project_Data.head(15)
```

```python
Project_Data.info()
```

```python
pr_da=Project_Data.drop(columns=["Unnamed: 0"])
```

```python
pr_da
```

```python
pr_da.columns =pr_da.iloc[0]
pr_da = pr_da[1:].reset_index(drop=True)
```

```python
pr_da.head(15)
```

```python
pr_da.info()
```

```python
pr_da["Date"] = pd.to_datetime(pr_da["Date"], dayfirst=True)
pr_da["Price"] = pd.to_numeric(pr_da["Price"])
pr_da["Quantity"] = pd.to_numeric(pr_da["Quantity"])
pr_da["Order ID"] = pr_da["Order ID"].astype(int)
```

```python
pr_da.info()
```

```python
pr_da.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
# Creating a Revenue column (Price * Quantity)
pr_da["Revenue"] = pr_da["Price"] * pr_da["Quantity"]
```

```python
pr_da.head(15)
```

```python
pr_da.drop_duplicates()
```

```python
#Total Revenue per Product 
pr_da.groupby("Product")["Revenue"].sum()
```

```python
# Average Quantity sold per Product
pr_da.groupby("Product")["Quantity"].mean()
```

```python
# Total Revenue per City
pr_da.groupby("City")["Revenue"].sum()
```

```python
# Average Revenue per Manager
pr_da.groupby("Manager")["Revenue"].mean()
```

```python
import matplotlib.pyplot as plt
```

```python
# Trend of Revenue over Time
revenue_trend = pr_da.groupby("Date")["Revenue"].sum()
```

```python
plt.plot(revenue_trend.index, revenue_trend.values)
plt.title("Revenue Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Quantities Sold
plt.hist(pr_da["Quantity"])
plt.title("Distribution of Quantities Sold")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Total Quantity Sold per Product 
product_quantity = pr_da.groupby("Product")["Quantity"].sum()
```

```python
product_quantity.plot(kind="bar", color="blue")
plt.title("Total Quantity Sold per Product")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```
